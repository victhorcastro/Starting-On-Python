# Codificação em Python

```python
# Estrutura Se-Então
...
if <condição>:
  <instruções para condição verdadeira>
...
```

```python
# Estrutura Se-Então-Senão
...
if <condição>:
  <instruções para condição verdadeira>
else:
  <instruções para condição falsa>
...
```

# Estruturas de Decisão Aninhadas

Podemos escrever estruturas de decisão dentro de outras estruturas de decisão, formando o que se chama de *Estrutura de Decisão Aninhada*.

```python
# Estrutura Condicional Aninhada
...
if <condição1>:
  if <condição2>:
    <instruções para ambas condições verdadeiras>
  else:
    <instruções para condição2 falsa>
else:
  <instruções para condição1 falsa>
...
```

# Controle de Fluxo
Usando instruções `if-elif-else`, podemos controlar o fluxo de um programa com base em diferentes condições e, então, executar o bloco de código apropriado de acordo com os resultados dos testes das condições.

```python
# Estrutura para controle de fluxo
...
if <condição1>:
  <instruções para condição1 verdadeira>
elif <condição2>:
  <instruções para condição1 verdadeira>
...
elif <condiçãoN>:
  <instruções para condiçãoN verdadeira>
else:
  <instruções para nenhuma das N condições verdadeiras>
...
```

##**Exercício 04**##

Elaborar um programa Python que permita a leitura de três números reais $a$, $b$ e $c$, calcule (se possível) e exiba as raízes reais da equação $ax^2+bx+c=0$. O algoritmo deve checar se a entrada de dados de fato representa uma equação do 2º grau.

```python
# Exercício 04
from math import sqrt

# Entrada de Dados
a = float(input('Digite o valor do parâmetro a: '))
b = float(input('Digite o valor do parâmetro b: '))
c = float(input('Digite o valor do parâmetro c: '))

# Checando se existe uma equação do 2º grau
if a != 0:
    #Cálculo do discriminante
    delta = b**2 - 4*a*c
    if delta >= 0:
        # Cálculo das raízes reais
        x1 = (-b + sqrt(delta))/(2*a)
        x2 = (-b - sqrt(delta))/(2*a)
        print(f'As raízes reais são {x1:.2f} e {x2:.2f}')
    else:
        print(f'O discriminante é negativo! Não há raízes reais!')
else:
    print('Não se trata de uma equação do 2º grau.')

```
